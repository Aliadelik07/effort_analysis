```python
import pandas as pd
import os

url = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vTc-X0GQYoyNgSJTYoTgXKJuaVOHIPTP9uVUXp5IMopOY_EQ6_KWKZnTUFLKEV3ymyaWdvUU6uutZX1/pub?gid=0&single=true&output=csv'

data = pd.read_csv(url)
output_directory = '/Users/ali/Desktop/Experiment/txt_files/'

# Drop rows with any NaN values
data = data.dropna(subset=['sub'])

for index, row in data.iterrows():
    # Extract values from the row
    sub = str(row['sub'])
    strategy = str(row['strategy'])
    test_train = str(row['test/train'])
    
    badchan_txt = str(row['_badchan.txt'])
    
    badica_txt = str(row['_bad_ica.txt'])
    
    reref_txt = str(row['_refchan.txt'])
    
    if badchan_txt != "-":
        # Create the file name
        file_name = f"{sub}-{strategy}-{test_train}_badchan.txt"

        # Write the content to the file
        with open(os.path.join(output_directory, file_name), 'w') as file:
            file.write(badchan_txt)
    
    if badica_txt != "-":
        # Create the file name
        file_name = f"{sub}-{strategy}-{test_train}_bad_ica.txt"

        # Write the content to the file
        with open(os.path.join(output_directory, file_name), 'w') as file:
            file.write(badica_txt)
    
    if reref_txt != "-":
        # Create the file name
        file_name = f"{sub}-{strategy}-{test_train}_refchan.txt"

        # Write the content to the file
        with open(os.path.join(output_directory, file_name), 'w') as file:
            file.write(reref_txt)
            


data
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

url = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vTc-X0GQYoyNgSJTYoTgXKJuaVOHIPTP9uVUXp5IMopOY_EQ6_KWKZnTUFLKEV3ymyaWdvUU6uutZX1/pub?gid=0&single=true&output=csv'

data = pd.read_csv(url)

# Extract the 'workingcapacity' column
working_capacity = data['workingcapacity']

# Propagate to 1000 points using interpolation
x_original = np.arange(len(working_capacity))
x_new = np.linspace(0, len(working_capacity) - 1, 1000)
working_capacity_interpolated = np.interp(x_new, x_original, working_capacity)

# Create the KDE plot
plt.figure(figsize=(10, 6))
sns.kdeplot(working_capacity_interpolated, shade=True)

# Add a vertical red line at 39
#plt.axvline(39, color='red', linestyle='--')

plt.title('Kernel Density Estimate of Working Capacity')
plt.xlabel('Working Capacity')
plt.ylabel('Density')
plt.show()
```
