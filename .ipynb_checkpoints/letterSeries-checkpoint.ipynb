```python
# rolling strategy

item = [] #rehears items
response = [] #subject's responses

#The first three trials
item.append(list_letters[0])
item.append(list_letters[1])
item.append(list_letters[2])

trial = 3
while trial < len(list_letters):
    
    #checking the item with the presented letter
    if item[0] == list_letters[trial]: 
        response.append(1)
    else:
        response.append(0)
    
    #updating the rehears items
    item[0] = item[1]
    item[1] = item[2]
    item[2] = list_letters[trial]
        

    #next trial
    trial += 1
    
rolling = response #saving response based on this strategy
len(rolling)
```

```python
#computing dprime
from scipy.stats import norm

strategies = [static, rolling, justPress, justDPress, grabNcount]
performances = []
scores = []

for response in strategies:
    
    #response = strategies[i]
    prob = 1 # realizatoin probability based on how effortful the strategy is: P(e).

    n_hit = 0.01
    n_fa = 0.01
    n_miss = 0.01
    n_cr = 0.01
    dprime = []
    score = []

    for i in range(len(response)):
        if response[i] == 1:
            if static[i] == 1:
                n_hit += 1
            else:
                n_fa += 1

        else:
            if static[i] == 1:
                n_miss += 1
            else:
                n_cr += 1


        dprime.append((norm.ppf(n_hit/(n_hit+n_miss))- norm.ppf(n_fa/(n_fa+n_cr)))*prob)
        score.append(n_hit - n_fa)

    performances.append(dprime)
    scores.append(dprime)
```
