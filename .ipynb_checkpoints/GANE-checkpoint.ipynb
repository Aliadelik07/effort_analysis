```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import hilbert, butter, filtfilt

# Define time parameters
t = np.linspace(0, 1, 1000)  # 1 second sampled at 1000 Hz

# Theta rhythm parameters
theta_freq = 5  # 5 Hz for theta rhythm

# Gamma rhythm parameters
gamma_freq1 = 30  # 30 Hz for first gamma rhythm
gamma_freq2 = 40  # 40 Hz for second gamma rhythm

# Generate signals
theta_signal = np.sin(2 * np.pi * theta_freq * t)
gamma_signal1 = np.sin(2 * np.pi * gamma_freq1 * t)
gamma_signal2 = np.sin(2 * np.pi * gamma_freq2 * t)

# Bandpass filter function
def bandpass_filter(signal, lowcut, highcut, fs, order=5):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype='band')
    y = filtfilt(b, a, signal)
    return y

# Compute phase-amplitude coupling using modulation index
def compute_modulation_index(phase, amplitude, num_bins=18):
    phase_bins = np.linspace(-np.pi, np.pi, num_bins + 1)
    amplitude_means = np.zeros(num_bins)

    for i in range(num_bins):
        indices = np.where((phase >= phase_bins[i]) & (phase < phase_bins[i + 1]))[0]
        if len(indices) > 0:
            amplitude_means[i] = np.mean(amplitude[indices])

    amplitude_means /= np.sum(amplitude_means)
    modulation_index = -np.sum(amplitude_means * np.log(amplitude_means + np.finfo(float).eps)) / np.log(num_bins)

    return modulation_index

# Sampling frequency
fs = 1000

# Initialize lists to store modulation indices and theta phases
modulation_indices = []
theta_phase_values = []

# Vary a_1 from 0 to 1
a_1_values = np.linspace(0, 1, 50)

for a_1 in a_1_values:
    a_2 = 1 - a_1

    # Adjust the amplitudes of the gamma signals
    adjusted_gamma_signal1 = a_1 * gamma_signal1
    adjusted_gamma_signal2 = a_2 * gamma_signal2

    # Interaction: sum of the signals
    resulting_signal = theta_signal + adjusted_gamma_signal1 + adjusted_gamma_signal2

    # Bandpass filter the resulting signal to isolate theta and gamma components
    theta_band = bandpass_filter(resulting_signal, 4, 8, fs)
    gamma_band = bandpass_filter(resulting_signal, 30, 50, fs)

    # Compute the analytic signal for theta and gamma bands
    theta_analytic = hilbert(theta_band)
    gamma_analytic = hilbert(gamma_band)

    # Extract the phase of the theta component
    theta_phase = np.angle(theta_analytic)
    
    # Append the mean theta phase to the list (convert to degrees for better interpretability)
    mean_theta_phase = np.mean(theta_phase)
    theta_phase_values.append(np.degrees(mean_theta_phase))

    # Extract the amplitude envelope of the gamma component
    gamma_amplitude = np.abs(gamma_analytic)

    # Compute the modulation index
    modulation_index = compute_modulation_index(theta_phase, gamma_amplitude)
    modulation_indices.append(modulation_index)

# Plot the modulation index as a function of a_1
plt.figure(figsize=(10, 6))
plt.plot(a_1_values, modulation_indices, marker='o')
plt.title('Phase-Amplitude Coupling (PAC) as a Function of $a_1$')
plt.xlabel('$a_1$')
plt.ylabel('Modulation Index')
plt.grid(True)

# Annotate each point with the corresponding mean theta phase (rounded to the nearest integer)
for i, txt in enumerate(theta_phase_values):
    plt.annotate(f'{int(round(txt/6))}°', (a_1_values[i], modulation_indices[i]), textcoords="offset points", xytext=(0, 10), ha='center')

# Show the plot
plt.show()

print("Completed plotting PAC as a function of a_1")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import hilbert, butter, filtfilt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from scipy import stats

# Define time parameters
t = np.linspace(0, 1, 1000)  # 1 second sampled at 1000 Hz

# Theta rhythm parameters
theta_freq = 5  # 5 Hz for theta rhythm

# Gamma rhythm parameters
gamma_freq1 = 30  # 30 Hz for first gamma rhythm
gamma_freq2 = 40  # 40 Hz for second gamma rhythm
gamma_freq3 = 50  # 50 Hz for third gamma rhythm

# Generate signals
theta_signal = np.sin(2 * np.pi * theta_freq * t)
gamma_signal1 = np.sin(2 * np.pi * gamma_freq1 * t)
gamma_signal2 = np.sin(2 * np.pi * gamma_freq2 * t)
gamma_signal3 = np.sin(2 * np.pi * gamma_freq3 * t)

# Bandpass filter function
def bandpass_filter(signal, lowcut, highcut, fs, order=5):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype='band')
    y = filtfilt(b, a, signal)
    return y

# Compute phase-amplitude coupling using modulation index
def compute_modulation_index(phase, amplitude, num_bins=18):
    phase_bins = np.linspace(-np.pi, np.pi, num_bins + 1)
    amplitude_means = np.zeros(num_bins)

    for i in range(num_bins):
        indices = np.where((phase >= phase_bins[i]) & (phase < phase_bins[i + 1]))[0]
        if len(indices) > 0:
            amplitude_means[i] = np.mean(amplitude[indices])

    amplitude_means /= np.sum(amplitude_means)
    modulation_index = -np.sum(amplitude_means * np.log(amplitude_means + np.finfo(float).eps)) / np.log(num_bins)

    return modulation_index

# Sampling frequency
fs = 1000

# Initialize list to store modulation indices
modulation_indices = []

# Vary a_1 from 0 to 1
a_1_values = np.linspace(0.5, 1, 10000)

for a_1 in a_1_values:
    a_2 = (1 - a_1) / 2

    # Adjust the amplitudes of the gamma signals
    adjusted_gamma_signal1 = a_1 * gamma_signal1
    adjusted_gamma_signal2 = a_2 * gamma_signal2
    adjusted_gamma_signal3 = a_2 * gamma_signal3

    # Interaction: sum of the signals
    resulting_signal = theta_signal + adjusted_gamma_signal1 + adjusted_gamma_signal2 + adjusted_gamma_signal3

    # Bandpass filter the resulting signal to isolate theta and gamma components
    theta_band = bandpass_filter(resulting_signal, 4, 8, fs)
    gamma_band = bandpass_filter(resulting_signal, 30, 60, fs)

    # Compute the analytic signal for theta and gamma bands
    theta_analytic = hilbert(theta_band)
    gamma_analytic = hilbert(gamma_band)

    # Extract the phase of the theta component
    theta_phase = np.angle(theta_analytic)

    # Extract the amplitude envelope of the gamma component
    gamma_amplitude = np.abs(gamma_analytic)

    # Compute the modulation index
    modulation_index = compute_modulation_index(theta_phase, gamma_amplitude)
    modulation_indices.append(modulation_index)

# Convert lists to numpy arrays for regression
a_1_values = np.array(a_1_values).reshape(-1, 1)
modulation_indices = np.array(modulation_indices)

# Polynomial regression of degree 2
poly = PolynomialFeatures(degree=1)
a_1_poly = poly.fit_transform(a_1_values)
model = LinearRegression().fit(a_1_poly, modulation_indices)

# Predict PAC values using the polynomial regression model
modulation_indices_pred = model.predict(a_1_poly)

# Calculate R-squared value
r_squared = r2_score(modulation_indices, modulation_indices_pred)

# Calculate p-value
n = len(a_1_values)
m = a_1_poly.shape[1]
y = modulation_indices
yhat = modulation_indices_pred
residuals = y - yhat
sse = np.sum(residuals**2, axis=0) / float(n - m)
se = np.sqrt(np.diagonal(sse * np.linalg.inv(np.dot(a_1_poly.T, a_1_poly))))
t = model.coef_ / se
p_values = [2 * (1 - stats.t.cdf(np.abs(i), (n - m))) for i in t]

# Print R-squared value and p-value
print(f"R-squared: {r_squared}")
print(f"P-values: {p_values}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import entropy

# Function to calculate the normalized Shannon entropy of a signal
def calculate_entropy(signal, num_bins=100):
    hist, bin_edges = np.histogram(signal, bins=num_bins, density=True)
    hist = hist + np.finfo(float).eps  # to avoid log(0)
    return entropy(hist)

# Generate oscillatory signals
t = np.linspace(0, 10, 1000)
signals = {
    'Sine wave 1 Hz': np.sin(2 * np.pi * 1 * t),
    'Sine wave 5 Hz': np.sin(2 * np.pi * 5 * t),
    'Sine wave 10 Hz': np.sin(2 * np.pi * 10 * t),
    'Sine wave 1 Hz + Noise': np.sin(2 * np.pi * 1 * t) + np.random.normal(0, 0.5, t.shape),
    'Combination of Sine waves': np.sin(2 * np.pi * 1 * t) + np.sin(2 * np.pi * 5 * t) + np.sin(2 * np.pi * 10 * t)
}

# Calculate entropy for each signal
entropies = {name: calculate_entropy(signal) for name, signal in signals.items()}

# Sort signals by entropy
sorted_signals = sorted(signals.items(), key=lambda item: entropies[item[0]])

# Plot the signals
plt.figure(figsize=(12, 8))
for i, (name, signal) in enumerate(sorted_signals, start=1):
    plt.subplot(len(signals), 1, i)
    plt.plot(t, signal)
    plt.title(f'{name} (Entropy: {entropies[name]:.4f})')
    plt.tight_layout()

plt.show()
```
