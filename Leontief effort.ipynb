```python
import matplotlib.pyplot as plt

# Define the offsets for e and K
e_offset = 0.4  # Increase in effort from e1 to e2
k_offset = 0.4  # Increase in cognitive strategy from K1 to K2

# Plot
fig, ax = plt.subplots()

# Lines and points for K1 and e1
ax.plot([0, 1.5], [0, 1.5], color="black",linestyle="--")  # Line with slope b/a
ax.plot([0.5, 1.5], [0.5, 0.5], color="black")  # Horizontal line at K1
ax.plot([0.5, 0.5], [0.5, 1.2], color="black")  # Vertical line at e1

# Points for K1 and e1
ax.plot(0.5, 0.5, 'ko')  # Point (e1, K1)
ax.plot(1.1, 0.5, 'ko')  # Point (e > e1, K1)
ax.plot(0.5, 0.9, 'ko')  # Point (e1, K > K1)

# Labels for K1 and e1
ax.text(0.45, 0.5, "(e₁, K₁)", ha="right")
ax.text(1.2, 0.4, "e > e₁", ha="right")
ax.text(0.45, 0.95, "K > K₁", ha="right")

# Lines and points for K2 and e2
ax.plot([0.5 + e_offset, 1.5 + e_offset], [0.5 + k_offset, 0.5 + k_offset], color="black", linestyle="-")  # Horizontal line at K2
ax.plot([0.5 + e_offset, 0.5 + e_offset], [0.5 + k_offset, 1.2 + k_offset], color="black", linestyle="-")  # Vertical line at e2

# Points for K2 and e2
ax.plot(0.5 + e_offset, 0.5 + k_offset, 'ko')  # Point (e2, K2)
ax.plot(1.1 + e_offset, 0.5 + k_offset, 'ko')  # Point (e > e2, K2)
ax.plot(0.5 + e_offset, 0.9 + k_offset, 'ko')  # Point (e2, K > K2)

# Labels for K2 and e2
ax.text(0.45 + e_offset, 0.5 + k_offset, "(e₂, K₂)", ha="right")
ax.text(1.2 + e_offset, 0.4 + k_offset, "e > e₂", ha="right")
ax.text(0.45 + e_offset, 0.95 + k_offset, "K > K₂", ha="right")

# Custom y-axis labels for discrete K values
ax.set_yticks([0.5, 0.5 + k_offset])
ax.set_yticklabels(["K₁", "K₂"])

# Custom x-axis labels for discrete e values
ax.set_xticks([0.5, 0.5 + e_offset])
ax.set_xticklabels(["e₁", "e₂"])

# Axes labels
ax.set_xlabel("Effort (e)")
ax.set_ylabel("Cognitive Strategy (K)")

# Limits and style
ax.set_xlim(0, 1.5 + e_offset)
ax.set_ylim(0, 1.5 + k_offset)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Save the plot
plt.savefig("/Users/ali/Desktop/Figures_/leontief_effort_cognitive_strategy_plot.png", format="png", dpi=300)  # Save as PNG with 300 DPI

plt.show()
```
